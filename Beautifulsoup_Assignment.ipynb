```python
# program to display IMDB’s Top rated 100 movies’ data

from bs4 import BeautifulSoup
import requests
import random
import re
import pandas as pd

url = 'https://www.imdb.com/chart/top'

response = requests.get(url)
soup = BeautifulSoup(response.text, 'lxml')
movies = soup.select('td.titleColumn')

 
ratings = [b.attrs.get('data-value')
           for b in soup.select('td.posterColumn span[name=ir]')]


#creating empty list
list = []


# Iterating over movies to extract
# each movie's details
for index in range(0, len(movies),):
     
   
    movie_string = movies[index].get_text()
    movie = (' '.join(movie_string.split()).replace('.', ''))
    movie_title = movie[len(str(index))+1:-7]
    year = re.search('\((.*?)\)', movie_string).group(1)
    data = {"Movie Title": movie_title,
            "Year": year,
            "Rating": (ratings[index])}
    list.append(data)
 
 # printing movie details with its rating.
for movie in list:
    
    df =  pd.DataFrame(list)
        
    
df.head(100)
```

```python
#program to scrape cricket rankings from icc-cricket.com - Men

from bs4 import BeautifulSoup as bs
import requests
import pandas as pd

url = "https://www.icc-cricket.com/rankings/mens/team-rankings/odi"
response = requests.get(url)

soup = bs(response.content, 'html.parser')
#Top 10 ODI teams in men’s cricket along with the records for matches, points and rating.

matches = []    #creating empty list for matches
points = []       #creating empty list of points of matches
ratings = []       #creating empty list  of ratings
teamName = []       #creating empty list of Team name
ranking = []         #creating empty list of ranking of team

team = soup.find_all('tr', class_='table-body')
topTeam = soup.find('tr', class_ ='rankings-block__banner')
rank =topTeam.find('td', class_='rankings-block__banner--pos')
ranking.append(rank.get_text())
td = topTeam.find('span', class_='u-hide-phablet')
teamName.append(td.get_text())

mNo = topTeam.find('td', class_='rankings-block__banner--matches')
matches.append(mNo.get_text())

mPoint = topTeam.find('td', class_='rankings-block__banner--points')
points.append(mPoint.get_text())

rate = topTeam.find('td', class_='rankings-block__banner--rating')
ratings.append(rate.get_text().replace('\n','').strip())

for i in team:
    
    td = i.find('span', class_='u-hide-phablet')
    teamName.append(td.get_text())
    
    mNo = i.find_all('td', class_='table-body__cell')
    matches.append(mNo[2].get_text())

    points.append(mNo[3].get_text())
    
    ratings.append(mNo[4].get_text())
    
    ranking.append(mNo[0].get_text())
    

#print(ranking)    
#print(teamName)    
#print(matches)    
#print(points)
#print(ratings)
print("  ***** TOP 10 ODI CRICKET TEAM *****  ")


#creating datafram of listes items
TopTeam = pd.DataFrame({'Rank':ranking, 'Team':teamName, 'Matches':matches, 'Points':points, 'Ratings':ratings})

TopTeam.head(10)
```

```python
#program to scrape details of all the posts from coreyms.com

from bs4 import BeautifulSoup
import requests
import re
import pandas as pd
import numpy as np
url = 'https://coreyms.com/'     #url for scraping
response = requests.get(url)
soup = BeautifulSoup(response.text, 'lxml')
#creating empty lists
heading = []
date = []
content = []
vdo = []


mainClass = soup.find_all('main' , class_=('content'))
for mainC in mainClass:
    for a in mainC.find_all('a', class_= ('entry-title-link')):   #coding for heading
        heading.append(a.get_text())
        
    for t in mainC.find_all('time', class_= ('entry-time')):      #coding for date
        date.append(t.get_text())
     
    for c in mainC.find_all('div', class_ = ('entry-content')):  #coding for content 
        p = c.find('p')
        content.append(p.get_text())
        
    for v in mainC.find_all('span', class_ =('embed-youtube')): 
        span = v.find('iframe', class_=('youtube-player'))
        vdo.append(span['src'], )
        
#print(len(heading))
#print(len(date))
#print(len(content))
#print(len(vdo))


df = pd.DataFrame.from_dict({'Heading': heading, 'Date':date, 'Content': content, 'Video':vdo}, orient= 'index')

df.transpose()
```

```python
#program to scrape mentioned details from dineout.co.in

from bs4 import BeautifulSoup as bs
import requests
import re
import pandas as pd
url = 'https://www.dineout.co.in/delhi-restaurants/buffet-special'     #url for scraping
response = requests.get(url)
soup = bs(response.text, 'lxml')
RestName = []
Cuisine = []
Location = []
Ratings = []
URL = []

RName= soup.find_all('div', class_='restnt-info cursor')
for i in RName:
    rName = i.find('a', class_='restnt-name ellipsis')
    RestName.append(rName.get_text())
    

    
location= soup.find_all('div', class_='restnt-loc')

for div in location:
    
    i = div.find_all('a')
    strloc=''
    for loc in i:
        strloc=strloc+' '+loc.get_text()
    
    Location.append(strloc.lstrip())
        
        
    
Cui= soup.find_all('span', class_='double-line-ellipsis')
for span in Cui:
    i = span.find_all('a')
    strcui=''
    for cui in i:
        strcui = strcui+ ', ' +cui.get_text()
    Cuisine.append(strcui.lstrip())
    
    
        
Rating=soup.find_all('div', class_='img-wrap')    
for i in Rating:
    rate = i.find('div', class_='restnt-rating rating-4')
    Ratings.append(rate.get_text())
    
    
url= soup.find_all('div', class_='img cursor')
for i in url:
    img1 = i.find('img', class_=('no-img'))
    URL.append(img1['data-src'],)
    
#print(RestName)  
#print(Location)
#print(Cuisine)
#print(Ratings)
#print(URL)

df =pd.DataFrame.from_dict({'Name':RestName, 'Cuisine':Cuisine, 'Location':Location, 'Ratings': Ratings, 'Img Url':URL}, orient='index')


df.transpose()
```

```python
#program to scrape weather details for last 24 hours from Tutiempo.net

from bs4 import BeautifulSoup as bs
import requests
import pandas as pd
from lxml import html
url = 'https://en.tutiempo.net/delhi.html?data=last-24-hours'
response = requests.get(url, verify=False)

soup = bs(response.content, 'html.parser')

Hour = []        #emply lists
Temperature = []
Wind = []
Weather = []
Humidity = []
Pressure = []

hour = soup.find_all('tr', class_='thHora')   # hour 
for i in hour:
    Hour.append(i.get_text())
    
print(Hour)  

temp = soup.find_all('td', class_='t Temp')     #temperature
for i in temp:
    Temperature.append(i.get_text())
    
#print(Temperature)  

wind= soup.find_all('td', class_='wind')        #wind
for i in wind:
    Wind.append(i.get_text())
    
print(Wind)   

humidity = soup.find_all('td', class_= 'hr')     #humidity
for i in humidity:
    Humidity.append(i.get_text())
    
print(Humidity) 

pressure = soup.find_all('td', class_='prob')      #pressure
for i in pressure:
    Pressure.append(i.get_text())
    
print(Pressure)  

wc = soup.find_all('span', class_='thhip ico i0530 u303n')     #weather
for i in wc:
    Weather.append(i.get_text())
    
print(Weather) 

df = pd.DataFrame.from_dict({'Hour':Hour,'Temperature':Temperature, 'Wind':Wind, 'Weather':Weather, 'Humidity':Humidity, 'Pressure':Pressure }, orient='index')

df.transpose()
```
